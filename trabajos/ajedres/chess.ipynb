```python
import chess
import random

board = chess.Board()
board
```

## Ejecución de movimientos aleatorios

```python
a = []
mov1 = 0

for mov in board.legal_moves:
    a.append(str(mov))
    tam = len(a)
    mov1 = random.randrange(tam)
    
print("Posibles movimientos:",a)
print("Movimiento ejecutado:",a[mov1])
board.push_san(a[mov1])
print("Es jaque Mate?:", board.is_checkmate())
board
```
